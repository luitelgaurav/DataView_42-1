### Radar Plot 

```python
categories = ['FGM', 'FGA', 'FG3M', 'FG3A',  'FTM', 'FTA']
categories_percent = ['FG_PCT' , 'FG3_PCT' , 'FT_PCT']
```

```python
selected_player_name='LeBron James'
team_selected_player=players[players['full_name']==selected_player_name]['team_id'].values[0]
selected_player=players[players['full_name']==selected_player_name]['id'].values[0]
```

```python
selected_player_stats=players_avg_stats[players_avg_stats['Player_ID']==selected_player][categories].values.flatten().tolist()
```

```python
team_selected_player_stats=teams_avg_stats[teams_avg_stats['Team_ID']==team_selected_player][categories].values.flatten().tolist()
```

```python
season_players_stats=players_stats.mean()[['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA',
       'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF','PTS']].round(2)#.values.flatten().tolist()
```

```python
import plotly.graph_objects as go


fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=selected_player_stats,
      theta=categories,
      fill='toself',
      name=selected_player_name
))
fig.add_trace(go.Scatterpolar(
      r=team_selected_player_stats,
      theta=categories,
      fill='toself',
      name='Team Stats'
))

fig.add_trace(go.Scatterpolar(
      r=season_players_stats[categories],
      theta=categories,
      fill='toself',
      name='Season Stats'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 22]
    )),
  showlegend=False
)

fig.show()
```

## Comparing Stats

```python
other_stats=['OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV','PF', 'PTS']
```

```python
arr_who=[]
arr_val=[]
arr_stat=[]
for col in other_stats:
    arr_who.append(selected_player_name)
    arr_stat.append(col)
    arr_val.append(players_avg_stats[players_avg_stats['Player_ID']==selected_player][col].values[0])

for col in other_stats:
    arr_who.append(teams[teams['id']==team_selected_player]['full_name'].values[0])
    arr_stat.append(col)
    arr_val.append(teams_avg_stats[teams_avg_stats['Team_ID']==team_selected_player][col].values[0])

for col in other_stats:
    arr_who.append("Season")
    arr_stat.append(col)
    arr_val.append(season_players_stats[col])
```

```python
comparison_df=pd.DataFrame([pd.Series(arr_who,name='Who'),pd.Series(arr_stat,name='Stat'),pd.Series(arr_val,name='Avg Value')]).T
comparison_df.head(2)
```

```python
fig = px.bar(comparison_df, y="Stat", x="Avg Value", 
                 color="Who", barmode="group" ,orientation='h' )
fig.update_layout(legend={'title_text':'','font_color':'white'},
                  paper_bgcolor='rgba(0,0,0,0)',
                  # plot_bgcolor='rgba(0,0,0,0)'
                 )
fig.update_xaxes(color='white')
fig.update_yaxes(color='white')
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
# fig.update_xaxes(title_font=dict(size=18, family='Courier', color='white'))
# fig.update_yaxes(title_font=dict(size=18, family='Courier', color='white'))
# Layout(
#     paper_bgcolor='rgba(0,0,0,0)',
#     plot_bgcolor='rgba(0,0,0,0)'
fig
```

## Exemplos

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# define data set
s1 = pd.Series(['dont have a mortgage',-31.8,'have mortgage',15.65])
s2 = pd.Series(['have utility bill arrears',-21.45,'',0])
s3 = pd.Series(['have interest only mortgage',-19.59,'',0])
s4 = pd.Series(['bank with challenger bank',-19.24,'bank with a traditional bank',32.71])
df = pd.DataFrame([list(s1),list(s2),list(s3),list(s4)], columns = ['label1','value1','label2','value2'])

# create subplots
fig = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=True,
                    shared_yaxes=True, horizontal_spacing=0)

fig.append_trace(go.Bar(y=df.index, x=df.value1, orientation='h', width=0.4, showlegend=False, marker_color='#4472c4'), 1, 1)
fig.append_trace(go.Bar(y=df.index, x=df.value2, orientation='h', width=0.4, showlegend=False, marker_color='#ed7d31'), 1, 2)
fig.update_yaxes(showticklabels=False) # hide all yticks
```

```python
fig = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_yaxes=True, horizontal_spacing=0)

# bar plot for tv shows
fig.append_trace(go.Bar(x=df.index, y=df.value1,
                        orientation='h', showlegend=True, 
                        # text=df, 
                        name='Value 1',
                        marker_color='#221f1f'), 1, 1)

# bar plot for movies
fig.append_trace(go.Bar(x=df.index, y=df.value2, 
                        orientation='h', showlegend=True,
                        # text=df_movie.rating,
                        name='Value 2', marker_color='#b20710'), 1, 2)

fig.show()
```
